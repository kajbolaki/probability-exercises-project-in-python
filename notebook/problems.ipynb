## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import random

total_trials = 1000
successful_trials = 0

for _ in range(total_trials):
    dice_1 = random.randint(1, 6)
    dice_2 = random.randint(1, 6)
    total_score = dice_1 + dice_2

    if total_score % 2 == 0 or total_score > 7:
        successful_trials += 1

estimated_probability = successful_trials / total_trials
print("Estimated probability:", estimated_probability)
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

total_trials = 1000
three_white_two_red = 0
same_color = 0

ball_values = list(ball_box.values())

for _ in range(total_trials):
    sample = random.choices(ball_values, k=5)

    white_count = sample.count("White")
    red_count = sample.count("Red")

    if white_count == 3 and red_count == 2:
        three_white_two_red += 1

    if len(set(sample)) == 1:
        same_color += 1

probability_three_white_two_red = three_white_two_red / total_trials
probability_same_color = same_color / total_trials

print("Probability of 3 white and 2 red:", probability_three_white_two_red)
print("Probability of all balls being the same color:", probability_same_color)
```
